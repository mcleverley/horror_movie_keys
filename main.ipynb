```python
# list_of_titles = ['The Shining', 'The Conjuring', 'Paranormal Activity']

list_of_titles = ['The Dark Knight', 'The Matrix', 'Kung Fu Panda']
```

```python
# uses list of title strings -> gets movie objects -> appends to list as objs
for movie_title in list_of_titles:
    list_movies = ia.search_movie(movie_title)
    tar = list_movies[0]
    movies_list.append(tar)

# objs don't have keywords by default, so update them all
for movie in movies_list:
    ia.update(movie, ['keywords'])
    
# now we have movies_list, a list of imdb.movie objects
# but you should think of them as dictionaries
# they have some cool functionality detailed in the IMDBpy documentation
# like the above, where we pulled keywords from each movie straight from IMDB w/ one line. neat
'''
potential issue:
do we include _all_ keywords? because some movies have much more keywords than others, like
The Dark Knight has 666 vs Kung Fu Panda at 162
so i'd like to assume the keywords are sorted by importance? we could compare the first 10 of each
movie vs the last 10 to check
'''
```

```python
for movie in movies_list[:1]:
    print(movie['title'])
    print('First 10')
    print(movie['keywords'][:10])
    print('Last 10')
    print(movie['keywords'][-10:])
    print('\n')
    
# yeah, this looks pretty good. the other movies generally follow it as well
```

```python
# so let's select the first... 30 keywords
for movie in movies_list:
    movie['keywords'] = movie['keywords'][:30]
```

```python
for movie in movies_list:
    print(len(movie['keywords'])) # nice
```

```python
import pandas as pd
from imdb import IMDb
```

```python
ia = IMDb() # make IMDb object. like a cursor or something
```

```python
df = pd.read_csv('movies.csv') # i found a list of 10k top rated movies on kaggle
# of course, we will not use this as our dataset
# it is merely here to assist us ;) 
```

```python
df.head(3) # this has... some stuff we need, but it is better to extract just the titles
```

```python
list_of_titles = df.Title.tolist()
```

```python
list_of_titles[:3]
```

```python
len(list_of_titles) # great
```

```python
# so now let's take our titles, and using the lovely IMDbPY library,
# we can get a list of 10,000 imdbpy.movie objects, which are dictionaries but fancier
```

```python
movies_list = []
```

```python
# this cell might take a while to load for some reason
for movie_title in list_of_titles:
    list_movies = ia.search_movie(movie_title) # search returns list of movies matching by relevance to search input
    tar = list_movies[0] # select closest match
    movies_list.append(tar)
    print(f'Appended {movie_title}')
    
print('done appending')


```

```python
print(len(movies_list))
```

```python
# the objs don't have keywords by default, so update them all
for movie in movies_list:
    ia.update(movie, ['keywords'])
```
